```python
# @author : Ayush Shyam Kumar
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=5 #slope
c=10 #intercept
x=np.array(list(range(-5,10))).T #Column Vector x
y=m*x+c #Coulmn Vector y
y=np.reshape(y,(len(x),1))
n=len(x) #number of data points
"""let us generate n disturbing values from normal distribution with sigma=5 . That is, standard deviation =5"""
noise= 5*np.random.randn(n,1) #nx1 column vector from normal distribution
yd=y+noise
A=np.column_stack((x,np.ones((n,1)))) #Appending a column of ones
Alpha=np.matmul(np.matmul(np.linalg.inv(np.matmul(A.T,A)),A.T),yd)
ycap=np.matmul(A,Alpha)
plt.plot(x,yd,'*') #Plotting scattered data points
plt.plot(x,ycap) #Plotting the regression line
plt.xlabel("'Independent Variable x")
plt.ylabel("Dependent Variable y")
plt.title("Simple Linear Regression")
```

```python
error=yd-ycap #Error Vector
check=np.matmul(A.T,error)
np.set_printoptions(suppress=True)
print(check) #Error vector perpendicular to Design Matrix
```
