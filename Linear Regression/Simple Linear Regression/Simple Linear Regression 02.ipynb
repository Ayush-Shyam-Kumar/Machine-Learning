```python
# @author : Ayush Shaym Kumar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the data set for simple linear regression
data=pd.read_csv(r"C:\Users\User\Desktop\Linear Regression\income.data.csv")
#plotting the data set for simple linear regression
plt.scatter(data['income'],data['happiness'],s=15)
plt.title("Income Vs Happiness")
plt.xlabel("Income")
plt.ylabel("Happiness")
```

```python
xdata=data['income'].to_numpy() #Independent variable
ydata=data['happiness'].to_numpy() #Dependent Variable
onesvector=np.ones(xdata.shape)
A=np.array([xdata,onesvector]).T
alpha=np.linalg.inv(A.T.dot(A)).dot(A.T).dot(ydata)
ycap=A.dot(alpha)
plt.scatter(xdata,ydata,label="datapoints",s=15) #Plotting scattered data points
plt.plot(xdata,ycap,c='r',label="Regression Line") #Plotting the regression line
plt.title("Income Vs Happiness")
plt.xlabel("Income")
plt.ylabel("Happiness")
plt.legend()
```
